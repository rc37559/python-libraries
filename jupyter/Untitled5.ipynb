```python
x=np.arange(0,10)
y=np.arange(11,21)
y=x**2

```

```python
# scatter plot
plt.scatter(x,y,c='g',marker="*")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("2-d plot")
plt.savefig("test.png")
```

```python
# plt plot
plt.plot(x,y,'g',linestyle="dashed")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("2-d plot")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,"b")
```
